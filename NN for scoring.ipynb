```python
# load x and y 
X = []
y = []
import csv
with open('history_data_bob.csv', newline='') as csvfile:
    history_data = csv.reader(csvfile, delimiter=',', quotechar='|')
    for sequence in history_data:
        X.append(sequence)
with open('score_data_bob.csv', newline='') as csvfile:
    score_data = csv.reader(csvfile, delimiter=',', quotechar='|')
    for score in score_data:
        y.append(score)
```
